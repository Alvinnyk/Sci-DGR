```python
import requests
import re
url = 'https://arxiv.org/pdf/2403.12043.pdf'
r = requests.get(url, stream=True)
name = re.findall(r'[^/]+(?=/$|$)', url)[0]
print(name)
with open(f'./papers/{name}', 'wb') as fd:
    for chunk in r.iter_content(2000):
        fd.write(chunk)
```

```python
import requests
import xml.etree.ElementTree as ET

# Define the search query
params = {
    'search_query': 'cat:physics',  # This searches all physics papers
    'start': 0,                     # Starting point of results
    'max_results': 5                # Number of results to return
}

# Send a GET request to the arXiv API
response = requests.get('http://export.arxiv.org/api/query', params=params)
print(response.content)
# Parse the XML response
root = ET.fromstring(response.content)
print(root)
# Iterate through the entries and print paper titles
for entry in root.findall('{http://www.w3.org/2005/Atom}entry'):
    print(entry)
    title = entry.find('{http://www.w3.org/2005/Atom}title').text
    print(title)
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://arxiv.org/list/astro-ph.GA/pastweek?show=107'

# Fetch the page
response = requests.get(url)
print(response)
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all div tags containing each paper's title
    # This selector might need adjustment based on the actual page structure
    titles = soup.find_all('div', class_='list-title mathjax')
    
    for title in titles:
        # Each title text is structured as "Title: actual_title_text"
        # So, we split on ": " and get the second part ([1])
        print(title.text.split(': ', 1)[1].strip())
else:
    print('Failed to retrieve the page')
```
